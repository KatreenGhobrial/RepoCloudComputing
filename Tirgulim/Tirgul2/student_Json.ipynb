```python
import json
import IPython
from IPython.display import display, HTML
import ipywidgets as widgets
from google.colab import drive
IPython.display.display_html("<b>Student Management System</b>", raw=True)

FILE_PATH = '/content/drive/My Drive/students.json'

out = widgets.Output()

student_dropdown = widgets.Dropdown(
    options=['Ester Peretz', 'Kathreen Ghobrial', 'Liel Etinger', 'Yohad Ben Baruch'],
    value='Yohad Ben Baruch',
    description='Student:',
)

link_text = widgets.Text(
    description='Program:',
    placeholder='Enter favorite program...'
)

update_button = widgets.Button(
    description='Update Program',
    button_style='success',
    icon='save'
)


def load_student_info(change):
    """Triggered when the dropdown value changes."""
    if change['name'] == 'value':
        out.clear_output()
        with out:
            try:
                with open(FILE_PATH, 'r', encoding='utf-8') as fid:
                    contents = json.load(fid)

                found = False
                for key in contents:
                    if contents[key].get('name') == change['new']:
                        current_val = contents[key].get('link', '')
                        link_text.value = current_val
                        print(f"--- Data for {change['new']} ---")
                        print(f"Current Program: {current_val if current_val else 'None'}")
                        found = True
                        break

                if not found:
                    print("Student not found in the JSON file.")
                    link_text.value = ""
            except Exception as e:
                print(f"Error reading file: {e}")

def save_changes(b):
    """Triggered when the update button is clicked."""
    selected_name = student_dropdown.value
    new_link = link_text.value

    with out:
        try:
            with open(FILE_PATH, 'r', encoding='utf-8') as fid:
                contents = json.load(fid)

            updated = False
            for key in contents:
                if contents[key].get('name') == selected_name:
                    contents[key]['link'] = new_link
                    updated = True
                    break

            if updated:
                with open(FILE_PATH, 'w', encoding='utf-8') as fid:
                    json.dump(contents, fid, indent=4, ensure_ascii=False)
                print(f" Success! {selected_name}'s program updated to: {new_link}")
            else:
                print(" Could not find student to update.")
        except Exception as e:
            print(f" Error saving to file: {e}")

student_dropdown.observe(load_student_info)
update_button.on_click(save_changes)

display(student_dropdown, link_text, update_button, out)
```
