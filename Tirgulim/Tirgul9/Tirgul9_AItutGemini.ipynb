Gemini: Creating a prompt
This rich example shows how you can create and configure complex prompts for Gemini. It assumes that you've already created an API key at https://aistudio.google.com/app/apikey and added it to your Colab secrets as GOOGLE_API_KEY (see the "Connecting to Gemini" snippet).

```python
import google.generativeai as genai

genai.configure(api_key="API-KEY")
model = genai.GenerativeModel("gemini-2.5-flash")
response = model.generate_content("You are an expert plant pathologist. A user reports that their indoor tomato plant has dark concentric spots on the lower leaves and the leaves are turning yellow. Based on common plant diseases, provide a short diagnosis of the most likely disease and 2 actionable steps to treat it using an IoT plant monitoring system.")
print(response.text)
```

```python
import nltk
from nltk.chat.util import Chat, reflections

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')

# Define some patterns and responses
patterns = [
 (r'hi|hello|hey', ['Hello! Welcome to the CloudRoot assistant.', 'Hi there! Ready to check on your plants?']),
    (r'how are you?', ['I\'m monitoring the sensors and doing great!', 'All systems are online and functioning perfectly.']),
    (r'what is your name?', ['I am the CloudRoot assistant.', 'You can call me CloudRoot.']),
    (r'(.*) disease|sick|spots|yellow', ['If you notice yellowing leaves or dark spots, our image recognition system can help identify the specific plant disease.']),
    (r'(.*) moisture|sensor|water', ['Our IoT sensors continuously track soil moisture to prevent conditions that lead to root rot and fungal diseases.']),
    (r'(.*) algorithm|model', ['We use convolutional neural networks (CNNs) to achieve high accuracy in detecting plant diseases from leaf images.']),
    (r'(.*) monitor|dashboard', ['You can view real-time data from your plant sensors directly on your personalized dashboard.'])]

# Create a chatbot
chatbot = Chat(patterns, reflections)

# Start the conversation
print("Hello! I'm Gemini. How can I help you today?")
while True:
    user_input = input("You: ")
    response = chatbot.respond(user_input)
    print("Gemini:", response)
```

```python
import pathlib
import textwrap
import google.generativeai as genai
from IPython.display import display
from IPython.display import Markdown

#Returns a Markdown object that can be used to render markdown from the modified text.
def  to_markdown(text):
  text = text.replace('•', ' *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
model = genai.GenerativeModel('gemini-2.5-flash')

# Display formatted Markdown response
response = model.generate_content("count number of chracters the following text: Collage emerged in 13th-century Europe, evolving through medieval & Victorian uses. Picasso and Braque's 1912 work popularized it, but earlier examples exist.")
to_markdown(response.text)
```
